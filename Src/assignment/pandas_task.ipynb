```python
import pandas as pd
import numpy as np

def automated_stat_analyzer(df, column_name):
    series = df[column_name].dropna()
    result = {}
    if pd.api.types.is_numeric_dtype(series):
        mean_var = series.mean()
        median_var = series.median()
        std_var = series.std()
        if mean_var > median_var:
            skewness = "Right Skewed"
        elif mean_var < median_var:
            skewness = "Left Skewed"
        else:
            skewness = "Approximately Symmetric"
        result = {
            "Mean": mean_var,
            "Median": median_var,
            "Standard Deviation": std_var,
            "Skewness": skewness
        }
    else:
        mode_var = series.mode()[0]
        result = {
            "Mode": mode_var
        }
    return result
#examples to try
#automated_stat_analyzer(df_test, "Sales_Amount")
#automated_stat_analyzer(df_test, "Customer_Age")
automated_stat_analyzer(df_test, "Product_Category") #bonus categorical
```

```python
def null_handling_strategy(df, strategy="fill_mean"):
    if strategy =='drop rows':
        return df.dropna()
    elif strategy =='fill mean':
        numeric_cols = df.select_dtypes(include=[np.number]).columns
        for col in numeric_cols:
            df = df.fillna({col: df[col].mean()})
        return df
    elif strategy == 'fill median':
        numeric_cols = df.select_dtypes(include=[np.number]).columns
        for col in numeric_cols:
            df = df.fillna({col: df[col].median()})
        return df
    elif df.isnull().sum().sum() == 0:
        return df
    else:
         raise ValueError('strategy must be drop rows or fill mean or fill median')
    return df
```
